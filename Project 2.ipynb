```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the title.basics file into a dataframetitles = pd.read_csv("title.basics.tsv", sep="\t")

# Read the title.ratings file into a dataframe
basics =  pd.read_csv("title.basics.tsv.gz", sep='\t', low_memory=False)
```

```python
# Filter for movies that were released in the United States
# Read the title-akas-us-only.csv file into a dataframe
akas = pd.read_csv("title-akas-us-only.csv")
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data files

ratings = pd.read_csv("title.ratings.tsv.gz", sep='\t', low_memory=False)

```

```python
# Merge dataframes
#data = pd.merge(titles, ratings, on="tconst")
#data = pd.concat([data, akas], axis=0)
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
filter_us_titles = basics['tconst'].isin(akas['titleId'])
basics = basics[filter_us_titles]
basics

```

```python
basics = basics.replace({'\\N':np.nan})
```

```python
ratings = ratings.replace({'\\N':np.nan})
```

```python
akas = akas.replace({'\\N':np.nan})
```

```python
# Filter dataframe
basics['startYear'] = basics['startYear'].astype(float)
```

```python
#keep startYear 2000-2022
basics = basics[(basics['startYear']>=2000)&(basics['startYear']<=2022)]
basics.head()
```

```python
#Eliminate movies that are null for runtimeMinute and genres
basics = basics.dropna(subset=['runtimeMinutes','genres'])
basics.isna().sum()
```

```python
#keep only titleType == movie
basics = basics.loc[ basics['titleType']=='movie']
```

```python
#Filtering out movies that are documentaries
filter_documentaries = basics['genres'].str.contains('documentary',case=False)
```

```python
#Using documentaries filter
basics = basics[~filter_documentaries]
```

```python
#Viewing info of basics
basics.info()
```

```python
#Filtering ratings based on basics
filter_basics = ratings['tconst'].isin(basics['tconst'])
ratings = ratings[filter_basics]
ratings
```

```python
basics.to_csv('Data/title-basics.csv', index=False)
```
